## Seleccionar renglones
#### loc[] vs iloc[]


```python
df.loc[[0,1,2]] #Poder acceder a los elementos a partir de su etiqueta: 0, 1 y 2
```

```python
df.iloc[0:3] #Acceder a traves de su posicion en lugar de su etiquta, [desde:hasta]
```
